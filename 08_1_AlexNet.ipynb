## **AlexNet Architecture**

AlexNet introduced key architectural innovations to handle the complexity of ImageNet data. It used larger convolution windows (e.g., 11×11 in the first layer) compared to LeNet to capture more visual details. Subsequent layers reduced window sizes (e.g., 5×5, 3×3) and added max-pooling layers (3×3 window, stride of 2). AlexNet had significantly more convolution channels than LeNet and included two fully connected layers with 4096 outputs, requiring nearly 1GB of model parameters. Due to limited GPU memory at the time, AlexNet originally employed a dual GPU design (half of the model on each GPU).

The network replaced sigmoid activations with ReLU for computational simplicity and more effective training. ReLU avoided vanishing gradients seen in sigmoid, enabling better parameter updates.

To reduce overfitting, AlexNet used dropout in the fully connected layers and extensive data augmentation techniques, such as flipping, clipping, and color adjustments. These strategies improved robustness and reduced the risk of overfitting, making AlexNet a pioneering model in deep learning.
