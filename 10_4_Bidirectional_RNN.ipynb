## **Bidirectional Recurrent Neural Networks**

a simple technique transforms any unidirectional RNN into a bidirectional
RNN. We simply implement two unidirectional RNN layers
chained together in opposite directions and acting on the same input. For
the first RNN layer, the first input is x1 and the last input is xùëá, but for the second RNN
layer, the first input is xùëá and the last input is x1. To produce the output of this bidirectional
RNN layer, we simply concatenate together the corresponding outputs of the two underlying
unidirectional RNN layers.
